<h3 align="left"><font color='#5E5273'>🔍 Score Average Rating</font></h3>

```python
def score_average_rating(likes,dislikes):
    if likes + dislikes == 0:
        return 0
    return likes / (likes+dislikes)
```

```python
df["score_average_rating"] = df.apply(lambda x: score_average_rating(x["likes"], x["dislikes"]), axis=1)
df.sort_values("score_average_rating", ascending=False).head(10)
```

<div style="border-radius:10px; border:#DEB887 solid; padding: 15px; background-color: #FFFAF0; font-size:100%; text-align:left">

<h3 align="left"><font color='#DEB887'>💡 Notes: </font></h3>
    
*  In the first scenario, the value is 0.6, and in the second scenario, it is 0.55, therefore the function will favor the scenario with 60%. However, if we have 2 up and 0 down, with a total of 2, and 100 up and 1 down, with a total of 101, the first scenario will yield 1, and the second scenario will yield 0.99. Consequently, the function will prioritize the first scenario, although the second scenario should win considering the higher frequency. In this method we have ratio but no frequency.

<div style="border-radius:10px; border:#D0C2F0 solid; padding: 15px; background-color: #FFF0F4; font-size:100%; text-align:left">

<h3 align="left"><font color='#5E5273'>🔍 Wilson Lower Bound</font></h3>

```python
def wilson_lower_bound(up,down,confidence=0.95):
    n = up + down
    if n == 0:
        return 0
    z = st.norm.ppf(1- (1 - confidence) / 2)
    phat = 1.0 * up / n
    return (phat + z * z / (2 * n) - z * math.sqrt((phat * (1 - phat) + z * z / (4 * n)) / n )) / (1 + z * z / n)
```

```python
df["wilson_lower_bound"] = df.apply(lambda x: wilson_lower_bound(x["likes"], x["dislikes"]), axis=1)
df.sort_values("wilson_lower_bound", ascending=False).head(10)
```

<div style="border-radius:10px; border:#DEB887 solid; padding: 15px; background-color: #FFFAF0; font-size:100%; text-align:left">

<h3 align="left"><font color='#DEB887'>💡 Notes: </font></h3>
    
* "Bernoulli's p parameter calculates the confidence interval and accepts the lower limit of this confidence interval as the WLB score. Bernoulli is a probability distribution that is used to calculate the probabilities of binary events, for example, the probability of getting heads in a coin flip.

* Let's say my upvote rate in the 600-400 scenario is 0.6. With the Wilson Lower Bound, we establish a confidence interval between 0.5 and 0.7 with a 95% probability. The Wilson Lower Bound selects the lower score range, which is 0.5 in this case. Thus, we obtain a more reliable result.

<h3 align="left"><font color='#5E5273'>🔍 Weighted Sorting Score </font></h3>

```python
view_score = MinMaxScaler(feature_range=(1,10)).fit_transform(df[["views"]])
likes_score = MinMaxScaler(feature_range=(1,10)).fit_transform(df[["likes"]])
dislike_score = MinMaxScaler(feature_range=(1,10)).fit_transform(df[["dislikes"]])
comment_count_score = MinMaxScaler(feature_range=(1,10)).fit_transform(df[["comment_count"]])
df["custom_weight"] = (view_score * 10/100 + likes_score * 40/100 + dislike_score * 30/100 + comment_count_score * 20/100)
df.sort_values(by="custom_weight",ascending=False).head(10)
```

<h3 align="left"><font color='#5E5273'>🔍 IMBD rating / sorting </font></h3>

```python
def weighted_rating(r, v, M, C):
    return (v/(v+M) * r ) + (M/(v+M) * C)
```

```python
average = (df["likes"] + df["dislikes"])/df.shape[0]
vote_count = df["likes"] + df["dislikes"]
M = 2500
C = average.mean()
df["degree"] = weighted_rating(average, vote_count, M, C)
df.sort_values(by="degree",ascending=False).head(10)
```
